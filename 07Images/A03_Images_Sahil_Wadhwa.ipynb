```python
## Next I want to check the RGB components of the image
## Here is the function used to make the histogram of the RGB distribution
def print_RGB_histogram(image_array):
    num_bins = 25

    fig, ax = plt.subplots(3, sharex=True)

    # the histogram of the data
    n, bins, patches = ax[0].hist(image_array[:,:,0].flatten(), num_bins, color="red")
    n, bins, patches = ax[1].hist(image_array[:,:,1].flatten(), num_bins, color="green")
    n, bins, patches = ax[2].hist(image_array[:,:,2].flatten(), num_bins, color="blue")

    plt.show()
```

```python
## printing the histogrma distribution of the RGB distribution
im_array = np.array(image)
print_RGB_histogram(im_array)
```

```python
## Finally I am going to make all the colors of the image look 2 times darker
im_array = np.array(image)
im_array_darkened = im_array * 0.5 ## this should darken the image quite a bit
im_array_darkened = im_array_darkened.astype('uint8') ## makes sure each pixel is between 0-255
imageDark = Image.fromarray(im_array_darkened) ## creating the new image
display(imageDark) ## displaying the new image
```

```python
## Finally I am going to make all the colors of the image look 2 times lighter
im_array = np.array(image)
im_array_lightened = im_array * 2 ## this should lighten the image quite a bit
im_array_lightened = im_array_lightened.astype('uint8') ## makes sure each pixel is between 0-255
imageLight = Image.fromarray(im_array_lightened) ## creating the new image
display(imageLight) ## displaying the new image
```

```python
## Conclusion
## The results are pretty shocking and my prediction was wrong as I thought the lightened image would look better
## The lightened image is almost unrecogninable as it seems the pixels were too close to 255
## The darkened image is suprisingly still visible as the face of the person and the spark got darker
## What I learned is that its easier to make all the colors closer to black than white even if its a lowlight image
```
